## Интерполяция кривой

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate as interp
import random
```

#### Генерация точек

```python
n = 15
x = random.sample(range(-20, 20), n)
y = random.sample(range(-20, 20), n)
f = interp.interp1d(x, y, kind='quadratic', fill_value='extrapolate')
```

#### Квадратичная интерполяция

```python
xnew = np.arange(-20, 20, 0.01)
ynew = f(xnew)
plt.plot(x, y, 'o', xnew, ynew, '-')
```

#### Интерполяция Лагранжа

```python
def langrange(xx, yy, point):
    s = 0
    for i in range(len(xx)):
        s += yy[i] * np.prod([
            (point - xx[j]) / (xx[i] - xx[j])
            for j in range(len(xx)) if j != i
        ])
    return s

sc_lagrange = (interp.lagrange(x, y))
```

```python
plt.plot(x, y, 'o', xnew, [langrange(x, y, point) for point in xnew], '-')
plt.show()
```

```python
plt.plot(x, y, 'o', xnew, sc_lagrange(xnew), '-')
plt.show()
```
